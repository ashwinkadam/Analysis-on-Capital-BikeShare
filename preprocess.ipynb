```python
import glob
csv_list = [f for f in glob.glob("*.csv")]
    
```

This is to get all the unique station IDs from all the CSVs.

Collect Station ID, corresponding Station Name


I am ignoring the longitude and latitude given in the CSV file as they are not unique with the station ID

Open each CSV

Take in Start Station ID and store the unique ones into a different dataframe

Take in Start Station Name and store the unique ones into a different dataframe

Take in End Station ID and store the unique ones into a different dataframe

Take in End Station Name and store the unique ones into a different dataframe


```python
def extract_start(df):
    df.dropna(subset=["start_station_id"], axis=0, inplace=True)
    df.drop_duplicates("start_station_id", inplace=True)
    a = list(df["start_station_id"])
    filt = list(map(lambda x: bool(digit_pattern.search(str(x))), a))

    df = df[filt]
    df["start_station_id"] = df["start_station_id"].astype("int32")
    df.drop_duplicates("start_station_id", inplace=True)

    return df

def extract_end(df):
    df.dropna(subset=["end_station_id"], axis=0, inplace=True)
    df.drop_duplicates("end_station_id", inplace=True)
    a = list(df["end_station_id"])
    filt = list(map(lambda x: bool(digit_pattern.search(str(x))), a))

    df = df[filt]
    df["end_station_id"] = df["end_station_id"].astype("int32")
    df.drop_duplicates("end_station_id", inplace=True)

    return df
```

```python
def extract_start2(df):
    df.dropna(subset=["Start station number"], axis=0, inplace=True)
    df.drop_duplicates("Start station number", inplace=True)
    a = list(df["Start station number"])
    filt = list(map(lambda x: bool(digit_pattern.search(str(x))), a))

    df = df[filt]
    df["Start station number"] = df["Start station number"].astype("int32")
    df.drop_duplicates("Start station number", inplace=True)

    return df 
    

def extract_end2(df):
    df.dropna(subset=["End station number"], axis=0, inplace=True)
    df.drop_duplicates("End station number", inplace=True)
    a = list(df["End station number"])
    filt = list(map(lambda x: bool(digit_pattern.search(str(x))), a))

    df = df[filt]
    df["End station number"] = df["End station number"].astype("int32")
    df.drop_duplicates("End station number", inplace=True)

    return df
```

```python
import re

digit_pattern = re.compile("^\d+\.*\d*$")
new_start_df = pd.DataFrame()
new_end_df = pd.DataFrame()

new_start_df2 = pd.DataFrame()
new_end_df2 = pd.DataFrame()

success_list = list()
error_list = list()

for f in csv_list:
    try:
        start_df = pd.read_csv(f, usecols=["start_station_id", "start_station_name", 
                                            "start_lat", "start_lng"])
        df = extract_start(start_df)

        new_start_df = new_start_df.append(df, ignore_index=True)
        new_start_df.drop_duplicates("start_station_id", inplace=True)

        l = [start_df]
        del start_df
        del l

        end_df = pd.read_csv(f, usecols=["end_station_id", "end_station_name",
                                            "end_lat", "end_lng"])
        df = extract_end(end_df)

        new_end_df = new_end_df.append(df, ignore_index=True)
        new_end_df.drop_duplicates("end_station_id", inplace=True)

        l = [end_df]
        del end_df
        del l

        success_list.append(f)

    except Exception as E:
        print(E)
        error_list.append(f)


for f in csv_list:
    try:
        start_df = pd.read_csv(f, usecols=["Start station number", "Start station"])
        df = extract_start2(start_df)

        new_start_df2 = new_start_df2.append(df, ignore_index=True)
        new_start_df2.drop_duplicates("Start station number", inplace=True)

        l = [start_df]
        del start_df
        del l

        end_df = pd.read_csv(f, usecols=["End station number", "End station"])
        df = extract_end2(end_df)

        new_end_df2 = new_end_df2.append(df, ignore_index=True)
        new_end_df2.drop_duplicates("End station number", inplace=True)

        l = [end_df]
        del end_df
        del l

        success_list.append(f)

    except Exception as E:
        error_list.append(f)
```
